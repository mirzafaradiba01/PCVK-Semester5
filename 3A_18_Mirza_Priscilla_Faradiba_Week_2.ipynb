```python
import cv2 as cv
import matplotlib.pyplot as plt

# Membaca gambar dalam warna asli
img = cv.imread('/content/drive/MyDrive/PCVK/kucing1.jpeg')

# Membuat citra baru hanya dengan saluran Red-Blue
red_blue = img.copy()
red_blue[:, :, 1] = 0  # Set saluran hijau menjadi 0 (hanya tinggalkan saluran merah dan biru)

# Menampilkan citra dalam saluran Red-Blue
plt.imshow(cv.cvtColor(red_blue, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Red-Blue Channel')
plt.show()
```

3. Tampilkan image baris ke 20-115, kolom 25-120!

```python
import cv2 as cv
import matplotlib.pyplot as plt

# Membaca gambar dalam warna asli
img = cv.imread('/content/drive/MyDrive/PCVK/kucing1.jpeg')

# Memilih bagian gambar yang ingin ditampilkan
roi = img[20:115, 25:120]

# Menampilkan bagian yang dipilih dari gambar
plt.imshow(cv.cvtColor(roi, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Region of Interest (ROI)')
plt.show()
```

4. Tampilkan image baris ke 5-30, semua kolom, channel Red saja!

```python
import cv2 as cv
import matplotlib.pyplot as plt

# Membaca gambar dalam warna asli
img = cv.imread('/content/drive/MyDrive/PCVK/kucing1.jpeg')

# Memilih bagian gambar yang ingin ditampilkan (baris ke-5 hingga ke-30 dan semua kolom)
roi = img[5:30, :]

# Mengambil saluran Red (merah) dari bagian yang dipilih
red_channel = roi[:, :, 2]

# Membuat gambar dengan saluran Red saja
red_image = cv.merge([red_channel, red_channel, red_channel])

# Menampilkan gambar dengan saluran Red saja
plt.imshow(cv.cvtColor(red_image, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Red Channel (Baris 5-30, Semua Kolom)')
plt.show()
```

5. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np

# Membaca gambar
img = cv.imread('/content/drive/MyDrive/PCVK/kucing1.jpeg')

# Membaca gambar asli
image_copy = img.copy()

# Membuat dan menambahkan kotak satu persatu
# Kotak 1 (hijau)
cv.rectangle(image_copy, pt1 = (30, 30), pt2=(80, 80), color=(0, 255, 0), thickness=7)

# Kotak 2 (merah)
cv.rectangle(image_copy, pt1=(150, 50), pt2=(100, 100), color=(0, 0, 255), thickness=5)

# Kotak 3 (biru)
cv.rectangle(image_copy, pt1=(100, 250), pt2=(150, 200), color=(255, 0, 0), thickness=6)

# Kotak 4 (kuning)
cv.rectangle(image_copy, pt1=(50, 250), pt2=(100, 200), color=(255, 255, 0), thickness=9)

# Kotak 5 (cyan)
cv.rectangle(image_copy, pt1=(100, 200), pt2=(50, 100), color=(0, 255, 255), thickness=5)

# Menampilkan gambar terakhir
plt.imshow(cv.cvtColor(image_copy, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Membalikkan gambar secara terbalik
flipped_image = cv.flip(image_copy, 0)

# Menampilkan gambar terakhir dalam posisi terbalik
cv2_imshow(flipped_image)
cv.waitKey(0)
cv.destroyAllWindows()
```
